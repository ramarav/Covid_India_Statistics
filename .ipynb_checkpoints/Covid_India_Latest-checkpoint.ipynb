```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
covid = pd.read_csv('https://data.humdata.org/hxlproxy/api/data-preview.csv?url=https%3A%2F%2Fraw.githubusercontent.com%2FCSSEGISandData%2FCOVID-19%2Fmaster%2Fcsse_covid_19_data%2Fcsse_covid_19_time_series%2Ftime_series_covid19_confirmed_global.csv&filename=time_series_covid19_confirmed_global.csv')
```

```python
covid
```

```python
covid.drop("Lat",axis=1,inplace=True)
covid.drop("Long",axis=1,inplace=True)
covid.drop("Province/State",axis=1,inplace=True)
```

```python
covid.to_csv('covid_humdata.csv')
```

```python
covid
```

```python
df = pd.read_csv("covid_humdata.csv", index_col='Country/Region') #set column as       index
update_df = df.T 
```

```python
df = update_df.drop('Unnamed: 0')
```

```python
df
```

```python
df.India.plot(legend=True) #plot usa column
plt.show()
plt.savefig("cases_cumulative_new.png")
```

```python
ndf=df.India
ndf
```

```python
difference = ndf.diff()
difference
```

```python
difference.plot.bar(legend=True) #plot usa column
plt.show()
plt.savefig("cases_daily_new.png")
```
